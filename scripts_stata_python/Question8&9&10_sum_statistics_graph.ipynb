```python
import os
import shutil
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from plotnine import *
import warnings

%matplotlib inline
warnings.filterwarnings("ignore")
```

```python
data_path = os.path.join("..", 'data', 'data_filtered_with_transformation.csv')
df = pd.read_csv(data_path)
print(df.head())
```

```python
# Summary statistics - Question 8
summary_stats = df[['newcomers', 'rents', 'popu', 'rents_per_capita']].describe()
print(summary_stats) # summary statistics
print(df.isnull().sum())
```

```python
# Create a graph of any type - Question 9
plot1 = (
    ggplot(df, aes(x='popu')) +
    geom_histogram(
        alpha=0.9,  
        binwidth=100,  
        color="white",
        fill="orange"  
    ) +
    labs(title="Distribution of Popularity", x="Popularity", y="Frequency")+ theme_bw()
)

print(plot1)
```

```python
# Save this graph
graph_folder_path = os.path.join("..", "graphs")
file_path = os.path.join(graph_folder_path, 'popularity_histogram.png')
plot1.save(filename=file_path, width=10, height=5, dpi=250)  
print(f"Plot saved at: {file_path}")
```
